# **Word Cloud**

Word Cloud'da en sık olan kelimeler büyük yazdırılıp, nadir olan kelimeler küçük yazdırılır.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
 
from google.colab import drive
drive.mount('/content/drive')
```

2011 yılında üniversite sıralamasında en çok hangi ülkenin kullanıldığının görselleştirilmesi yapılmak amaçlanmaktadır.

```python
timesData = pd.read_csv("/content/drive/MyDrive/data/timesData.csv")
```

```python
timesData.head()
```

```python
x2011 = timesData.country[timesData.year == 2011] #2011 yılı filtre olarak seçilir, datanın ülkelerine bu filtre uygulanır.
```

```python
x2011.value_counts()  #oluşturulacak plotun en çok United States'ten oluşması beklenir.
```

```python
plt.subplots(figsize=(8,8))  #figür oluşturulur ve boyutu belirlenir.
wordcloud = WordCloud(       #wordcloud kütüphanesinden WordCloud methodu çağrılır.
                          background_color='white',  #arkaplan rengi belirlenir.
                          width=512,
                          height=384
                         ).generate(" ".join(x2011)) #kullanılan kelimeler ayrılıp büyük bir şekilde plot edilir.
plt.imshow(wordcloud)    #imshow -> image show demektir, bir image plot ettirilmek istenildiğinde kullanılır.
plt.axis('off')   #bir resim plot ettirileceğinden x ve y eksenlerinin gözükmemesi için off denilir.

plt.show()
```
