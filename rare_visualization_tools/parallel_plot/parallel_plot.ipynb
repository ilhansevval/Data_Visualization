# **Parallel Plot**

Parallel plot bir dizi sayısal değişken üzerinde birçok bireysel gözlemin featurelarını karşılaştırmaya izin verir.

Her dikey çubuk, bir feature'ı temsil eder.

Bu plotta iris dataset kullanılacaktır.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates

from google.colab import drive
drive.mount('/content/drive')
```

```python
data = pd.read_csv('/content/drive/MyDrive/data/iris.csv') 
```

```python
data = data.drop(['Id'],axis=1) #görselleştirmede gerek olmadığı için id sütunu datadan çıkartılır.
                                #axis=1 demek yukarıdan aşağı yani sütun drop et demektir.
```

```python
data.head() #4 feature ve türler vardır.
```

```python
data.Species.unique() #3 adet tür var.
```

```python
plt.figure(figsize=(15,10)) #figür boyutu belirlenerek figür oluşturuldu.
parallel_coordinates(data, 'Species', colormap=plt.get_cmap("Set1"))
#parelel coordinates methodunda;
#data ile Species sütununun görselleştireleceği belirtildi.
#colormap olarak Set1 isimli renk paleti kullanılacaktır. Bu renk paleti turuncu, kırmızı, gri renk içerir.
plt.title("Iris data class visualization according to features (setosa, versicolor, virginica)")
plt.xlabel("Features of data set")
plt.ylabel("cm")
plt.show()
```

Burada featureların türler arası değişimi görülmek istenmektedir.

Setosa türü PetalLengthCm'e göre diğer türlerden rahat bir şekilde ayrılabilir. Aynı zamanda versicolor ve virginica'nın da birbirinden gözle görülür bir şekilde ayrıldığı görülür. Buradan PetalLengthCm sınıflandırma yapmak için önemli bir feature olabilir denilebilir.


SepalWidthCm ise sınıflandırma için önemli değildir çünkü sınıflar çok karışıktır, türler birbirinden ayırt edilemez.
