# **Spyder Chart**

Radar plot da denilir.

Çok değişkenli verileri üç veya daha fazla nicel değişkenden oluşan iki boyutlu bir çizelge biçiminde görüntülemenin grafiksel bir yöntemidir.

```python
import pandas as pd 
import matplotlib.pyplot as plt
from math import pi

from google.colab import drive
drive.mount('/content/drive')
```

```python
data = pd.read_csv('/content/drive/MyDrive/data/iris.csv') 
```

```python
data = data.drop(['Id'],axis=1) #görselleştirmede gerek olmadığı için id sütunu datadan çıkartılır.
                                #axis=1 demek yukarıdan aşağı yani sütun drop et demektir.
```

```python
categories = list(data)[:4]
N = len(categories)
angles = [ n / float(N)*2*pi for n in range(N)]
angles = angles + angles[:1]
plt.figure(figsize = (10,10))
ax = plt.subplot(111,polar = True)
ax.set_theta_offset(pi/2)
ax.set_theta_direction(-1)
plt.xticks(angles[:-1],categories)
ax.set_rlabel_position(0)
plt.yticks([0,2,4,6],["0","2","4","6"],color= "red", size = 7)
plt.ylim(0,6)

values = data.loc[0].drop("Species").values.flatten().tolist()
values = values + values[:1]
ax.plot(angles,values,linewidth = 1,linestyle="solid",label ="setosa" )
ax.fill(angles,values,"b",alpha=0.1)

values = data.loc[1].drop("Species").values.flatten().tolist()
values = values + values[:1]
ax.plot(angles,values,linewidth = 1,linestyle="solid",label ="versicolor" )
ax.fill(angles,values,"orange",alpha=0.1)
plt.legend(loc = "upper left",bbox_to_anchor = (0.1,0.1))
plt.show()
```

Setosa ve versicolor çiçeğin türleridir. Bu plotta iki class ele alınmıştır.

Daire etrafında da 4 tane feature var. Plota bakıldığında featureların classlara göre değişimi görülmektedir.
