```python
timesData = pd.read_csv("/content/drive/MyDrive/data/timesData.csv")
```

```python
timesData.head()
```

```python
dataframe = timesData[timesData.year == 2015]
```

```python
trace1 = go.Scatter(
    x=dataframe.world_rank,  #x eksenlerine dünya sıralaması koyulur.
    y=dataframe.research,
    name = "research"      #ilk trace'e domain atamaya gerek yoktur, diğer tracelerin domain'i vardır.
)                          #domain ile plotların yeri belirlenir.
trace2 = go.Scatter(
    x=dataframe.world_rank,
    y=dataframe.citations,
    xaxis='x2',
    yaxis='y2',
    name = "citations"
)
trace3 = go.Scatter(
    x=dataframe.world_rank,
    y=dataframe.student_staff_ratio,
    xaxis='x3',
    yaxis='y3',
    name = "student_staff_ratio"
)
trace4 = go.Scatter(
    x=dataframe.world_rank,
    y=dataframe.teaching,
    xaxis='x4',
    yaxis='y4',
    name = "teaching"
)
data = [trace1, trace2, trace3, trace4] #trace1, trace2, trace3 ve trace4 data değişkenine liste olarak atanır.
layout = go.Layout(
    xaxis=dict(
        domain=[0, 0.45]  #plotların konumları teker teker belirlenir.
    ),
    yaxis=dict(
        domain=[0, 0.45]
    ),
    xaxis2=dict(
        domain=[0.55, 1]
    ),
    xaxis3=dict(
        domain=[0, 0.45],
        anchor='y3'
    ),
    xaxis4=dict(
        domain=[0.55, 1],
        anchor='y4'
    ),
    yaxis2=dict(
        domain=[0, 0.45],
        anchor='x2'
    ),
    yaxis3=dict(
        domain=[0.55, 1]
    ),
    yaxis4=dict(
        domain=[0.55, 1],
        anchor='x4'
    ),
    title = 'Research, citation, income and student staff ratio VS World Rank of Universities' #başlık oluşturulur.
)
fig = go.Figure(data=data, layout=layout) #data parametresine data değişkeni, layout parametresine layout değişkeni atanarak figür oluşturulur.

iplot(fig) #figür plot edilir.
```
