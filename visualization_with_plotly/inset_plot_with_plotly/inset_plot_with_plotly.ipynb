# **Inset Plot**

Bu plotta iki plot bir çerçevede bulunur.

Vurgulanmak istenen plot büyük kısımda, vurgulanmak istenmeyen yardımcı plot da küçük kısımda yer alır.

Bu plotta üniversitelerin teaching scoreları ve incomeları karşılaştırılmak amaçlanmaktadır.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from plotly.offline import iplot
import plotly.graph_objs as go

from google.colab import drive
drive.mount('/content/drive')
```

```python
timesData = pd.read_csv("/content/drive/MyDrive/data/timesData.csv")
```

```python
timesData.head()
```

```python
dataframe = timesData[timesData.year == 2015]
```

```python
# first line plot
trace1 = go.Scatter(  #scatter plot seçilir.
    x=dataframe.world_rank, #x eksenine üniversitelerin dünya sıralaması koyulur.
    y=dataframe.teaching,  #y eksenine teaching scoreları koyulur.
    name = "teaching",     #plotun yanındaki labeldır.
    marker = dict(color = 'rgba(16, 112, 2, 0.8)'), #renk ve saydamlık belirlenir.
)
# second line plot
trace2 = go.Scatter(  #scatter plot seçilir.
    x=dataframe.world_rank, #x eksenine üniversitelerin dünya sıralaması koyulur.
    y=dataframe.citations,  #y eksenine alıntı yapma değeri koyulur.
    xaxis='x2',             
    yaxis='y2',             #ana plotun içerisinde küçük plot yapılmak için kullanılır.
    name = "citations",     #plotun yanındaki labeldır.
    marker = dict(color = 'rgba(160, 112, 20, 0.8)'), #renk ve saydamlık belirlenir.
)

data = [trace1, trace2]  #trace1 ve trace2 data değişkenine liste olarak atanır.

layout = go.Layout(
    xaxis2=dict(
        domain=[0.6, 0.95],  #2. plotun yeri belirlenir.
        anchor='y2',         #2 plot olduğu için anchor kullanılır.
    ),
    yaxis2=dict(
        domain=[0.6, 0.95],  #2. plotun yeri belirlenir.
        anchor='x2',
    ),
    title = 'Citations and Teaching vs World Rank of Universities' #başlık verildi.

)

fig = go.Figure(data=data, layout=layout) #data parametresine data değişkeni, layout parametresine layout değişkeni atanarak figür oluşturulur

iplot(fig)
```
